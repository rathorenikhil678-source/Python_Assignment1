```python
import pandas as pd
```

```python
#DataFrame from two-dimensional list
data = [[35,"nikhil",21],
        [36,"omm",20],
        [37,"paarth",20]
        ]
df = pd.DataFrame(data,columns=["id","name","age"])

print(df)
```

```python
#DataFrame from Dictionary
data = {
        "name":["nikhil","om","paarth"],
        "age":[21,20,22]
        }

df= pd.DataFrame(data)
print(df)
```

```python
#DataFrame from List of Lists
data=[
    ["Nikhil",20],
    ["Rahul",21],
    ["Om",22]
    ]

df=pd.DataFrame(
                data,
                columns=["Name","Age"]
                )

print(df)
```

```python
#DataFrame from List of Tuples
data=[
    ("Nikhil",20),
    ("Rahul",21),
    ("Om",22)
]

df=pd.DataFrame(
    data,
    columns=["Name","Age"]
)

print(df)
```

```python
#DataFrame from List of Dictionaries
data=[
    {"Name":"Nikhil","Age":20},
    {"Name":"om","Age":21},
    {"Name":"paarth","Age":22}
]

df=pd.DataFrame(data)

print(df)
```
