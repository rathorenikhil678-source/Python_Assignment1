```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "ID":[1,2],
    "Name":["Nikhil","Rahul"]
})

df2 = pd.DataFrame({
    "ID":[3,4],
    "Name":["Om","Vikas"]
})

df3 = pd.DataFrame({
    "ID":[1,2,3,4],
    "Marks":[80,90,70,95]
})



combined = pd.concat([df1,df2])

print("Concatenated Data:")
print(combined)



result = pd.merge(
    combined,
    df3,
    on="ID"
)


print(result)
```

pd.concat() : combines rows
pd.merge() :joins using common column
