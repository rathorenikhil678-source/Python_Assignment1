```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "ID":[1,2,3],
    "Name":["Nikhil","Rahul","Om"]
})

df2 = pd.DataFrame({
    "ID":[1,2,4],
    "City":["Jaipur","Delhi","Mumbai"]
})

print(df1)

print(df2)


```

```python
# Inner Merge
inner = pd.merge(df1,df2,on="ID",how="inner")


print(inner)


# Left Join
left = pd.merge(df1,df2,on="ID",how="left")


print(left)


# Right Join
right = pd.merge(df1,df2,on="ID",how="right")


print(right)


# Index-based Join
join_result = df1.join(
    df2.set_index("ID"),
    on="ID"
)


print(join_result)
```

inner → only matching IDs
left → keeps all rows from left DataFrame
right → keeps all rows from right DataFrame
Missing values become NaN
join() uses index
