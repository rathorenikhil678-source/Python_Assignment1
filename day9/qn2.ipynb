```python
import pandas as pd

dates = pd.Series([
    "2025-08-01",
    "2025-09-10",
    "2025-10-15"
])

dates = pd.to_datetime(dates)


print(dates)

print("\nYear")
print(dates.dt.year)


print(dates.dt.month)

#omly days
print(dates.dt.day)

print(dates.dt.day_name())

#month name
print(dates.dt.month_name())
```
