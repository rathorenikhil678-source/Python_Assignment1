```python
import pandas as pd

df = pd.read_csv("loan_data.csv")

print(df.head())
```

```python
print(df.info())


print(df.shape)

print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Fill missing values

df["past_due_days"] = df["past_due_days"].fillna(0)

df["paid_off_time"] = df["paid_off_time"].fillna("Not Paid")

print(df.isnull().sum())
```

```python
df["effective_date"] = pd.to_datetime(
    df["effective_date"]
)


print(df.dtypes)
```

```python
print(
    "Average Principal:",
    df["Principal"].mean()
)
```

```python
print(
    df["loan_status"].value_counts()
)
```

```python
print(
    df.groupby("Gender")["age"].mean()
)
```
